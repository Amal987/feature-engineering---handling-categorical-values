### for multiple category

dataset can be taken from : https://www.kaggle.com/competitions/mercedes-benz-greener-manufacturing/data?select=train.csv.zip

```python
data = pd.read_csv('mercedesbenz.csv')
```

```python
data.head()
```

```python
print(f"dataset has {data.shape[0]} rows and {data.shape[1]} columns")
```

```python
len(data['X0'].unique())
```

```python
def column_unique_count(data):
    for col in data.columns:
        print(col,"---->", len(data[col].unique()))
```

```python
column_unique_count(data)
```

lets more frequent category column from the dataset

```python
df = data[['X0','X1','X5','X6','X8']]
```

```python
df.head()
```

```python
column_unique_count(df)
```

```python
df['X0'].value_counts()
```

here we have 47 different categories in 'X0' feature. If we do One hot encoding here, it will cause curse of dimensionality. so lets take 'n' most frequent category from the column and perform remaining steps

```python
# lets take n as 10
df['X0'].value_counts().sort_values(ascending = False).head(10)
```

```python
top_10_X0 = [x for x in df['X0'].value_counts().sort_values(ascending = False).head(10).index]
```

```python
top_10_X0
```

we only take these 10 category to perform one hot encoding and remaining categories will be dropped

```python
for category in top_10_X0:
    df[category] = np.where(df['X0']==category,1,0)
```

```python
df.head(10)
```

here we can see that, values belonging to top_10_X0 get 1 else 0

#### like this all features can be done.

```python
def encoding(data,feature,top_10):
    for category in top_10:
        data[category] = np.where(data[feature]==category,1,0)
```

```python
top_10_X1 = [x for x in df['X1'].value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10_X1
```

```python
encoding(df,'X1',top_10_X1)
```

```python
df.head()
```

and then we can drop the actual feature

```python
#for finding the top 10 category i append the code below

def top_10(data,feature):
    category = []
    category = [x for x in data[feature].value_counts().sort_values(ascending=False).head(10).index]
    return category
```

```python
top_1_X5 = top_10(df,'X5')
top_1_X6 = top_10(df,'X6')
top_1_X8 = top_10(df,'X8')
```

```python
print("X5-->",top_1_X5)
print("X6-->",top_1_X6)
print("X8-->",top_1_X8)
```
